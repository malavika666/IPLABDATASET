```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('flower8.jpg' )
plt.imshow(img)
plt.show()
max_channels = np.amax([np.amax(img[:,:,0]), np.amax(img[:,:,1]),np.amax(img[:,:,2])])

print(max_channels)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('i7.jpg' )
plt.imshow(img)
plt.show()
min_channels = np.amin([np.min(img[:,:,0]), np.amin(img[:,:,1]),np.amin(img[:,:,2])])

print(min_channels)
```
