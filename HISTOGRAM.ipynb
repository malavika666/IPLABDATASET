```python
#numpy.ndarray.flatten() function return a copy of the array collapsed into one dimension.
#hist() function in pyplot module of matplotlib library is used to plot a histogram
#Python numpy cumsum() function returns the cumulative sum of the elements along the given axis.
# xlim() returns the x-axis limits of the current plot
    
    
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

#Reading an image from file
path = "C:/Users/User/Desktop/RESEARCH ARTICLES/low contrast.jpeg"
img = cv.imread(path)

#Displaying the image
cv.imshow('Image',img)
k = cv.waitKey(0)
if k==27 or k==ord('q'):
    cv.destroyAllWindows()
    
# To view histogram
hist, bins = np.histogram(img.flatten(),256,[0,256])     
cdf = hist.cumsum()          
cdf_normalized = cdf*float(hist.max())/cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(img.flatten(), 256, [0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('Cdf', 'Histogram'),loc = 'upper left')
plt.show()

#Equalization
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
equ = cv.equalizeHist(gray)

#Displaying image
cv.imshow('equ.png', equ)
k = cv.waitKey(0)
if k == 27 or k == ord('q'):
    cv.destroyAllWindows()
    
#To view histogram
hist, bins = np.histogram(equ.flatten(),256,[0,256])      
cdf = hist.cumsum()          
cdf_normalized = cdf*float(hist.max())/cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(equ.flatten(), 256, [0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('Cdf', 'Histogram'),loc = 'upper left')
plt.show()




```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt 

#Reading an image from file
img = cv2.imread("C:/Users/User/Desktop/RESEARCH ARTICLES/low contrast.jpeg")

#Displaying the image
cv2.imshow('Image',img)
k = cv2.waitKey(0)
if k==27 or k==ord('q'):
    cv2.destroyAllWindows()

#Equalization    

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
equ = cv2.equalizeHist(gray)

#Displaying the altered image
cv2.imshow('equ.png', equ)
k = cv2.waitKey(0)
if k == 27 or k == ord('q'):
    cv2.destroyAllWindows()

#Plotting the Histogram
plt.hist(img.flat, bins=100, range=(0, 255))
plt.legend(('Actual Image', 'Histogram'),loc = 'upper left')
plt.show()
plt.hist(equ.flat, bins=100, range=(0, 255))
plt.legend(('Altered Image', 'Histogram'),loc = 'upper left')
plt.show()


    
```

```python
#CLAHE does histogram equalization in small patches or in small tiles with high accuracy and contrast limiting.

import cv2
import numpy as np
from matplotlib import pyplot as plt 

img = cv2.imread("C:/Users/User/Desktop/RESEARCH ARTICLES/test.png")

clahe = cv2.createCLAHE(clipLimit =2.0, tileGridSize=(8,8))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cl_img = clahe.apply(gray)


plt.hist(cl_img.flat, bins=100, range=(100, 255))
#plt.legend(('Actual Image', 'Histogram'),loc = 'best')
plt.show()

#ret, thresh1 = cv2.threshold(cl_img, 190, 150, cv2.THRESH_BINARY)
#ret, thresh2 = cv2.threshold(cl_img, 190, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(cl_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

#Displaying the altered image
cv2.imshow('test.png', thresh3)
k = cv2.waitKey(0)
if k == 27 or k == ord('q'):
    cv2.destroyAllWindows()


```
